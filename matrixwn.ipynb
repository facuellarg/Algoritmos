importar librerias

```python
from time import time
import matplotlib.pyplot as plt
```

se declaran algunas variables

```python
n = 10
TAM = 50
tiempos = []
```

funcion que calcula el tiempo promedio

```python
while n < TAM:

	m1 =[[1 for x in range(n)] for y in range(n)]
	m2 =[[2 for x in range(n)] for y in range(n)]
	mr =[[0 for x in range(n)] for y in range(n)]
	t = 0
	res = 0
	while t < 10:
		t = t + 1
		i = 0
		ti = time()
		while i < n:
			j = 0
			while j < n:
				k = 0
				while k < n:
					mr[i][j] = mr[i][j] + (m1[i][k] * m2[k][j])
					k = k+1
				j=j+1

			i=i+1
		tf = time()
		res = res + tf - ti
	tiempos.append(res/10)
	n = n + 1
```

se grafica el arreglo co los tiempos

```python
plt.plot(tiempos)
```
