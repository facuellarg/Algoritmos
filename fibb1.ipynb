```python
from time import time
import matplotlib.pyplot as plt
```

importacion de librerias


```python

def fibInt(n):
	n1 = 0
	n2 = 1
	a = 0
	ti = time()
	if n == 0 or n == 1:
		tf = time()
		return (tf - ti)
	while (a <  n - 1):
		n2 = n2 + n1
		n1 = n2 - n1
		a = a + 1
	tf = time()

	return (tf - ti)
```

definicion iterativa de la funcion fibb

```python
l = range(200)		
lt = []
```

```python
ti = time()
for x in l:
	lt.append(fibInt(x))
```
