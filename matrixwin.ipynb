importar las librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

se declaran algunas variables

```python
n = 10
TAM = 50
tiempos = []
```

funcion para calcular el tiempo 

```python
while n < TAM:
	m1 = np.ones((n,n))
	m2 = m1+1
	mr = np.zeros((n,n))
	t = 0
	res = 0
	while t < 10:
		t = t + 1
		i = 0
		ti = time()
		while i < n:
			j = 0
			while j < n:
				k = 0
				while k < n:
					mr[i][j] = mr[i][j] + (m1[i][k] * m2[k][j])
					k = k+1
				j=j+1

			i=i+1
		tf = time()
		res = res + tf - ti
	tiempos.append(res/10)
	n = n + 1
```

se grafica los tiempos guardados

```python
plt.plot(tiempos)
```
